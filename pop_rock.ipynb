```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
pop_rock = pd.read_csv('pop_rock.csv')
pop_rock
```

```python
pop_rock.describe()
```

```python
d=pd.DataFrame({'sd': np.std(pop_rock),
                'mean': np.mean(pop_rock)
               })
d
```

```python
d['three_sd_plus']=d['mean'] + 3*d['sd']
d['three_sd_minus']=d['mean'] - 3*d['sd']
d
```

```python
new_df = pop_rock.drop(['artist_names','artists_id','release_date','song_title (censored)', 'Genre'],axis = 1)
for column in d.index:
    new_df = new_df[(new_df[column] > d.loc[column, 'three_sd_minus']) & (new_df[column] < d.loc[column, 'three_sd_plus'])]
new_df
```

```python
colormap=plt.cm.RdBu
plt.figure(figsize=(14,12))
sns.heatmap(new_df.drop(columns=['Unnamed: 0','Unnamed: 0.1','year']).astype(float).corr(),linewidths=0.1,cmap=colormap,vmax=1.0,square=True,annot=True)
```

```python
fig,ax=plt.subplots(figsize=(15,7))
sns.lineplot(x='energy',y='loudness',data=new_df,ax=ax)
```

```python
new_df = new_df.assign(decade = lambda x: (x.year//10)*10)
new_df = new_df.drop(['Unnamed: 0','Unnamed: 0.1'],axis = 1)
new_df
```

```python
twenty = new_df[(new_df["decade"] == 1920)]
thirty = new_df[(new_df["decade"] == 1930)]
fourty = new_df[(new_df["decade"] == 1940)]
fifty = new_df[(new_df["decade"] == 1950)]
sixty = new_df[(new_df["decade"] == 1960)]
seventy = new_df[(new_df["decade"] == 1970)]
eighty = new_df[(new_df["decade"] == 1980)]
ninety = new_df[(new_df["decade"] == 1990)]
two = new_df[(new_df["decade"] == 2000)]
twoten = new_df[(new_df["decade"] == 2010)]
twotwenty = new_df[(new_df["decade"] == 2020)]
```
