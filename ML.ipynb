# Logistic Regression

```python
LR = pd.read_csv('two_thousand.csv')
LR.head()
```

```python
LR = LR.drop(['artist_names','key','mode','explicit','year','song_title','popularity','decade','Genre'],axis = 1)
LR['duration_ms'] = LR['duration_ms']/1000
LR.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
LR_classifier = LogisticRegression(solver='lbfgs',
   max_iter=100000,
   random_state=42)
```

```python
LR_classifier.fit(X_train, y_train)
```

```python
y_pred = LR_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
LR_accuracy=accuracy_score(y_test, y_pred)
```

```python
print("Accuracy: " + str(LR_accuracy))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
matrix = confusion_matrix(y_test, y_pred)
print(matrix)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

# DecisionTree

```python
# Initial imports
import pandas as pd
from path import Path
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Define the features set.
X = genre_binary_encoded.copy()
X = X.drop(["is_popular"], axis=1)
X.head()
```

```python
# Define the target set.
y = genre_binary_encoded["is_popular"].values
y[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Determine the shape of our training and testing sets.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Creating the decision tree classifier instance.
dt_model = tree.DecisionTreeClassifier()
# Fitting the model.
dt_model = dt_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = dt_model.predict(X_test_scaled)
predictions
```

```python
# Calculating the confusion matrix
dt_cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
dt_cm_df = pd.DataFrame(
    dt_cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

dt_cm_df
```

```python
# Calculating the accuracy score.
dt_acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(dt_cm_df)
print("Accuracy: " + str(dt_acc_score))
print("Classification Report")
print(classification_report(y_test, predictions))
```

# RandomForestClassifier

```python
# Initial imports.
import pandas as pd
from path import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Define the features set.
X = genre_binary_encoded.copy()
X = X.drop(["is_popular"], axis=1)
X.head()
```

```python
# Define the target set.
y = genre_binary_encoded["is_popular"].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
rf_predictions = rf_model.predict(X_test_scaled)
rf_predictions
```

```python
# Calculating the confusion matrix.
rf_cm = confusion_matrix(y_test, rf_predictions)

# Create a DataFrame from the confusion matrix.
rf_cm_df = pd.DataFrame(
    rf_cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

rf_cm_df
```

```python
# Calculating the accuracy score.
rf_acc_score = accuracy_score(y_test, rf_predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(rf_cm_df)
print("Accuracy: " + str(rf_acc_score))
print("Classification Report")
print(classification_report(y_test, rf_predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```

```python
# We can sort the features by their importance.
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```

# Boosting

```python
# Define the features set.
X = genre_binary_encoded.copy()
X = X.drop(["is_popular"], axis=1)
X.head()
```

```python
y = genre_binary_encoded["is_popular"].ravel()
y[:5]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Create a classifier object
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    classifier = GradientBoostingClassifier(n_estimators=20,
                                            learning_rate=learning_rate,
                                            max_features=5,
                                            max_depth=3,
                                            random_state=0)

    # Fit the model
    classifier.fit(X_train_scaled, y_train)
    print("Learning rate: ", learning_rate)

    # Score the model
    print("Accuracy score (training): {0:.3f}".format(
        classifier.score(
            X_train_scaled,
            y_train)))
    print("Accuracy score (validation): {0:.3f}".format(
        classifier.score(
            X_test_scaled,
            y_test)))
    print()
```

```python
boost_classifier = GradientBoostingClassifier(n_estimators=20,
   learning_rate=0.75, max_features=5, max_depth=3, random_state=0)

boost_classifier.fit(X_train_scaled, y_train)
boost_predictions = boost_classifier.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
boost_acc_score = accuracy_score(y_test, boost_predictions)
print(f"Accuracy Score : {boost_acc_score}")
```

```python
boost_cm = confusion_matrix(y_test, boost_predictions)
boost_cm_df = pd.DataFrame(
   boost_cm, index=["Actual 0", "Actual 1"],
   columns=["Predicted 0", "Predicted 1"]
)
display(boost_cm_df)
```

```python
print("Classification Report")
print(classification_report(y_test, boost_predictions))
```
