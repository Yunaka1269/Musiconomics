```python
# Create a SQLite database

from pathlib import Path
Path('spotify_music.db').touch()
```

```python
# Create a SQLite table

import sqlite3
conn = sqlite3.connect('spotify.db')
c = conn.cursor()
```

```python
# Execute query to create music_genre table and associated columns

c.execute('''CREATE TABLE music (song_id int, 
                                 artist_names text,
                                 artists_id int,
                                 danceability int,
                                 energy int,
                                 valence int,
                                 tempo int,
                                 loudness int,
                                 mode int,
                                 key int,
                                 acousticness int,
                                 instrumentalness int,
                                 liveness int,
                                 speechiness int,
                                 explicit int,
                                 duration_ms int,
                                 popularity int,
                                 year date,
                                 release_date date,
                                 song_title text,
                                 genre text)''')
```
